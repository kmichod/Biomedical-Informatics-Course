```python
# Use the below datasets create any four visualization summarizing the datasets. You can use either Matplotlib, Seaborn or any other python based visualization package. Be creative!. Here are some example visualizations: https://blog.hubspot.com/marketing/types-of-graphs-for-data-visualization. 
# Remember to label and have legends for your graphs. Add comments to describe what you are trying to graph
import pandas as pd
import seaborn as sb
import matplotlib as plot
diabetic_data = pd.read_csv("/Users/kayleighlawson-michod/Desktop/dataset_diabetes/diabetic_data.csv")
```

```python
diabetic_data
```

```python
# Visualization 1, stacked bar graph of insulin use distribution by gender
insulin_plt = diabetic_data.groupby(['gender', 'insulin']). size(). unstack(). plot(kind='bar', stacked=True)
insulin_plt.set_xlabel(r"Sex")
insulin_plt.set_ylabel(r"Count")
insulin_plt.set_title("Distribution of insulin by gender")
```

```python
# Visualization 2, bar graph of average time in hospital by sex
time_hospital = diabetic_data.groupby(['gender'])['time_in_hospital'].mean().plot(kind='bar')
time_hospital.set_xlabel(r"Sex")
time_hospital.set_ylabel(r"Days in hospital")
time_hospital.set_title("Average time in hosptial by gender")
```

```python
# Visualization 3, distribution of insulin in all patients

insulin = diabetic_data.groupby(['insulin']).size().plot(kind='pie')
insulin.set_title("Distribution of insulin in all patients")
```

```python
# Visualization 4, horizontal bar graph of DM medicaiton use by race or ethnicity
dm_plot = diabetic_data.groupby(['race', 'diabetesMed']).size().unstack().plot(kind='barh')
dm_plot.set_xlabel(r"Count")
dm_plot.set_ylabel(r"Reported race or ethnicity")
dm_plot.set_title("Diabetees medication use by reported race or ethnicity")
```
